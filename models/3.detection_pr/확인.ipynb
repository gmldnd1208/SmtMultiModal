```python
import pandas as pd
from ultralytics import YOLO

# model = YOLO("runs/detect/train/weights/best.pt")

metrics = model.val(
    data="./data.yaml",
    split="val",
    plots=True
)

cm = metrics.confusion_matrix
matrix = cm.matrix

names = list(model.names.values())
labels = names + ["background"]

df = pd.DataFrame(
    matrix,
    index=[f"pred_{name}" for name in labels],
    columns=[f"true_{name}" for name in labels]
)

print(df)
```
